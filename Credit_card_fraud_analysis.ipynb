```python
# Check the outliers
Q1 = df['Amount'].quantile(0.25)
Q2 = df['Amount'].quantile(0.75)

IQR = Q2 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q2 + 1.5*IQR

print(f"Q1: {Q1:.2f} and Q2: {Q2:.2f}")
print(f"Outliers bound: {lower_bound:.2f} - {upper_bound:.2f}")
print(f"Outliers count: {((df['Amount'] < lower_bound) | (df['Amount'] > upper_bound)).sum()}")

#boxplot
plt.figure(figsize=(10,5))
sns.boxplot(y = df['Amount'])
plt.axhline(y = upper_bound, color = 'red', linestyle = '--')
plt.ylim(0,500)
plt.title("Outlier visible\n(Real banking view)")
plt.show()
```

```python
outlier = df['Amount'] > upper_bound
print(f"Bank alert: {df[outlier]['Amount'].count()} greater than {upper_bound:.2f}+")
print(f"Fraud concentration: {df[outlier]['Class'].mean()*100:.2f}% v/s 0.17% avg in total")
```

```python
#Now compare normal transaction with fraud transaction

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
fraud_amt = df[df['Class']==1]['Amount']
normal_amount = df[df['Class']==0]['Amount'].sample(n = len(fraud_amt)*3, random_state = 42)
plt.hist(normal_amount, bins = 30, alpha = 0.7, color = 'green')
plt.hist(fraud_amt, bins = 30, alpha = 0.7, color = 'red')
plt.title("Amount: Fraud v/s Normal")

plt.subplot(1,2,2)
k = df.groupby('Class')['Amount'].median()
k.plot(kind = 'bar', color = ['green','red'])
plt.title('Median amount by class')
plt.ylabel('₹')
plt.xticks([0,1],['Normal','Fraud'],rotation = 0)
plt.show()
```

```python
df['Hour'] = (df['Time']//3600) % 24

plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
df['Hour'].hist(bins = 24, alpha = 0.7, color = 'blue')
plt.title("All transactions by hours")
plt.xlabel("Hour (0-23)")

plt.subplot(1,2,2)
df.groupby(df['Hour'])['Class'].mean().plot(kind = 'bar')
plt.title("Fraud rate by Hour")
plt.ylabel('Fraud %')
plt.xlabel('Hour')
plt.xticks(rotation = 0)
plt.show()
```

Above plots show that Most of the transactions happen during day. However, there are some transactions which happens at night and if compare with fraud rate by hours, those also happens at night. This shows that most of the fraud transactions happen at night.

```python
# Find features that best separate fraud/normal
top_features = ['V14','V17','V12','V10','V16']
fig, axes = plt.subplots(2,3,figsize = (15,10))
axes = axes.flatten()

for i, feature in enumerate(top_features):
    df.boxplot(column = feature, by = 'Class', ax = axes[i])
    axes[i].set_title(f'{feature}: Fraud vs Normal')

plt.tight_layout()
plt.show()
```
